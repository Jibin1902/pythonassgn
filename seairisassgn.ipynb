```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris=sns.load_dataset('iris')

df=pd.DataFrame(iris)
df.head()
```

```python
sns.histplot(iris['sepal_length'],bins=10,kde=True)
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.show()
```

```python
#boxplot

sns.boxplot(x='species',y='sepal_length',data=iris)
plt.show()
```

```python
#violinplot

sns.violinplot(x='species',y='sepal_length',data=iris)
plt.show()
```

```python
#swarmplot

sns.swarmplot(x='species',y='sepal_length',data=iris)
plt.show()
```

```python
#Stripplot

sns.stripplot(x='species',y='sepal_length',data=iris)
plt.show()
```

```python
#barplot
sns.barplot(x='species',y='sepal_length',data=iris,width=0.2)
plt.show()
```

```python
#countplot
sns.countplot(x='species',data=iris,width=0.2)
plt.show()
```

```python
#heatmap
numerical_df=df.select_dtypes(include=['float64','int64'])
sns.heatmap(numerical_df.corr(),annot=True)

plt.show()
```

```python
#scatterplot
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.show()
```

```python
#lineplot
sns.lineplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.show()
```
